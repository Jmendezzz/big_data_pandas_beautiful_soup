```python
import requests 
```

```python
DATA_URL = "https://www.worldometers.info/world-population/world-population-by-year/"
```

```python
data = requests.get(DATA_URL)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.content)
```

```python
tables = soup.find_all('table')
```

```python
import pandas as pd 
```

```python
df = pd.read_html(str(tables))
df = df[0]
df = df.fillna(0)
```

```python
df.to_csv(r"data|")
```
