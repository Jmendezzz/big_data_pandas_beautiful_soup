```python
import pandas as pd
```

```python
df = pd.read_csv("heart.csv")
```

```python
print("Punto 1.1")
df.head()
```

```python
print("Punto 1.2: Estadisticas Descriptivas")
df.describe()
```

```python
print("Punto 2.1: Eliminar fila 1")
df.drop(1).head()
```

```python
print("Punto 2.2: Eliminar columna 'ca'")
df.drop("ca", axis=1).head()
```

```python
print("Punto 2.3: Eliminar columna 'cp'")
df.drop("cp", axis=1).head()
```

```python
print("Punto 3.1: Selección con loc")
print(df.loc[[0, 2, 4, 6, 8, 10]])
print(df.loc[:,["age","sex"]])
```

```python
print("Punto 3.2: Selección con iloc:")
print(df.iloc[3:9])
```

```python
print("Punto 3.3: Selección de columnas por nombre:")
print(df[["age", "sex", "target"]].head())
```

```python
print("Punto 3.4: Selección de columnas por posición:")
print(df.iloc[:, [0, 1, -1]].head())
```

## Que diferencia hay entre loc e iloc?
loc permite la seleccion a traves de etiquetas y ubicacion de los elementos, en cambio iloc se basa en el indice de los eleementos por eso al seleccionar df.iloc[3:9] solo se selecciona desde el 3 hasta el 8

```python
print("Punto 4.1: Ordenar por colesterol:")
print(df.sort_values(by="chol", ascending = False).head())
```

```python
print("Punto 4.2: Promedios por sexo:")
print(df.groupby("sex").mean())
```

```python
print("Punto 4.3: Conteo por condición cardíaca:")
print(df.groupby("target").count())
```
